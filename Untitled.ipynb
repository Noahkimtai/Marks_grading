```python
#match all the quiz files to be merged
quizFiles=glob.glob('quiz_*_grades.csv')
quizFiles
```

```python
#concatinate all the seperate files into a single dataframe
#use the email address as the index column
allQuizes=pd.concat((pd.read_csv(i,converters={'Email':str.lower}, usecols=['Email','Grade',],index_col=['Email'])
                   for i in quizFiles),axis=1)
allQuizes.columns=['quiz 1','quiz 2','quiz 3','quiz 4','quiz 5']
allQuizes.head()
```

```python
#read the roster file
roster = pd.read_csv('roster.csv',
                    converters={'NetID':str.lower,'Email Address':str.lower},
                    usecols=['Section','Email Address','NetID'],
                    index_col='NetID')
roster.head()
```

```python
#read the homework file
homeworkExams=pd.read_csv('hw_exam_grades.csv',
                         converters={'SID':str.lower},
                         usecols= lambda x:'Submission' not in x,
                          index_col='SID')
homeworkExams.head()
#here the converter is used even if on inspection all colunms are in lower case for consistency
#usecols accepts one line function i.e the column is includde if it returns true
```

```python
#merge the homework and roster files together
rosterHomework=pd.merge( roster,homeworkExams,left_index=True, right_index=True,)
rosterHomework.head()
```

```python
#merge the quiz filea and the merged homework & roster file
allMarks=pd.merge(rosterHomework,allQuizes,left_on= 'Email Address',right_index=True)
#to check all the columns included
allMarks.columns
```

```python
allMarks.head()
```

```python
#fill null data
allMarks.fillna(0)
```

```python
#the different tests are assigned different weights towards the finel grade
weights=pd.DataFrame(data=[['Exam 1 Score' ,5,  0.05],
                          ['Exam 2 Score' ,10,  0.10],
                          ['Exam 3 Score',15,  0.15],
                          ['Quiz Score',  30,  0.30],
                          ['Homework Score' ,40,  0.40]],
                    columns=['category','Percent of Final Grad','Weight'])
weights
```

```python
#to calculate the score of the three exams between 0.0 & 1.0
exams=3
for i in range (1, exams+1):
    allMarks[f'Exam {i} Score']=(allMarks[f'Exam {i}']/allMarks[f'Exam {i} - Max Points'])
    
allMarks

```

```python
#select the maximum homework 
#select and calculate the total homework score
#use regular expression, axis parameter is included to operate on the colunns
#axis=0 would select rows matching the regex operation
homeworkScore=allMarks.filter(regex=r"^Homework \d\d?$", axis=1)
homeworkMaxScore=allMarks.filter(regex=r"^Homework \d\d? -", axis=1)
```

```python
#compute the sum of the selected coluns
totalHomework=homeworkScore.sum(axis=1)
totalMaxhw=homeworkMaxScore.sum(axis=1)
#by default sum() adds the column values but using axis=1 adds tells it to add row values
allMarks['Homework Total Score']=totalHomework/totalMaxhw
allMarks.columns
```

```python
allMarks['Homework Total Score'].head()
```

```python
#another approach is to calculate the ratio for each individual hoework and add them later
#This approach favours students who performed consistently
```

```python
#calculating the quiz score
#create  dataframe series to hold the quiz maximum score
quizMaxScore=pd.Series({'Quiz 1':11, 'Quiz 2':15, 'Quiz 3':17, 'Quiz 4':14, 'Quiz 5':12})
#use regex to match quiz score
quizScore=allMarks.filter(regex=r'^quiz \d\d?$',axis=1)
quizScoreSum=quizScore.sum(axis=1)
quizMaxScoreSum=quizMaxScore.sum()
allMarks['Total Quiz Score']=quizScoreSum/quizMaxScoreSum
allMarks['Total Quiz Score'].head()
```

```python
#calculating letter grade
#first applying the weights
scoreWeights=pd.Series(
                        {'Exam 1 Score' :  0.05,
                        'Exam 2 Score' :  0.10,
                        'Exam 3 Score' :  0.15,
                        'Total Quiz Score' :  0.30,
                        'Homework Total Score' :  0.40})
```

```python
allMarks['Final Score']= (allMarks[scoreWeights.index]*scoreWeights).sum(axis=1)
allMarks['Ceiling Score']=np.ceil(allMarks['Final Score']*100) #numpy.ceil() is used to round up the score after multiplying by 100
allMarks['Ceiling Score'].head()
```

```python
#map each score to a grade A-F
#use a dictionary
grades={
        90: 'A',
        80: 'B',
        70: 'C',
        60: 'D',
        0: 'F'
       }

def gradeMapping(value):
    for key, letter in grades.items():
        if value >= key:
            return letter
```

```python
#use Series.map() function to apply the gradeMapping function to the marks
#it applies an arbitrary function to values in series
#loop oover the values comparing the values to the key in the dictionary
#if the value fall in the bracket you return the grade
letterGrades = allMarks['Ceiling Score'].map(gradeMapping)
allMarks['Final Grade'] = pd.Categorical(
                                        letterGrades, categories=grades.values(), ordered=True
                                        )
allMarks['Final Grade'].head()
```

```python
#since the data was from different classes (Section) you need to group them and sort
#Use groupby() and sort() functions
allMarks.groupby('Section')
allMarks.sort_values(by = ['Last Name','First Name'])
    
```

```python
#plot grades to show distribution
gradeCounts = allMarks['Final Grade'].value_counts().sort_index()
gradeCounts.plot.bar() # you can use .plot.hist() to plot an histogram istead
plt.show()
```
